### 1. Getting libraries and loading data

```python
import pandas as pd
import missingno as msno
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("movie_metadata.csv")
data.head(5)
```

```python
col_supp = ['facenumber_in_poster', 'movie_imdb_link', 'aspect_ratio']   #Drop irrelevent columns
data.drop(col_supp, axis=1, inplace=True)
```

### 2. Remove duplicate movies

```python
x=data.movie_title.unique().shape
y=data.movie_title.shape
x,y
```

126 duplicate movie names

```python
data.movie_title.value_counts().head()
```

```python
data.drop_duplicates(subset="movie_title", inplace=True)
```

```python
msno.bar(data)
```

## 3.Dealing with missing values

### a. Director and budget

```python
data.drop(data[data.director_name.isna()].index, inplace=True)
```

```python
new_budg = round(data.budget.mean())
data.loc[data.budget.isna(), 'budget'] = new_budg
```

### b. Colour and language

```python
data.loc[data.color.isna(), 'title_year']
```

```python
data.color.fillna('Color', inplace=True)
```

```python
data.language.unique()
```

```python
data.loc[data.language == 'None', 'country']
```

```python
data.loc[data.language.isna(), 'country']
```

```python
data.language.fillna('English', inplace=True)
data.language.replace('None', 'English', inplace=True)
```

### c. Duration and title year

```python
data.loc[data.duration.isna()].shape
```

```python
data.drop(data.loc[data.duration.isna()].index, inplace=True)
```

```python
data.loc[data.title_year.isna()]
```

```python
data.drop(data.loc[data.title_year.isna()].index, inplace=True)
```

### d. Country and content rating

```python
data.loc[data.country.isna(),('movie_title','actor_1_name','actor_2_name','actor_3_name','director_name')]
```

```python
data.loc[(data.actor_1_name == 'Chris Brochu')|
         (data.actor_2_name == 'Jeff Fahey')  |
         (data.actor_3_name == 'Rita Wilson'),'country']
```

```python
data.country.fillna('USA', inplace=True)
```

```python
data.content_rating.unique()
```

```python
conv = {'PG':'PG', 'PG-13':'PG-13', 'G':'PG', 'R':'R', np.nan:'Unrated', 'Unrated':'Unrated',
       'TV-14':'PG-13', 'Not Rated':'Unrated', 'Approved':'Approved', 'NC-17':'NC-17',
       'X':'NC-17', 'GP':'PG', 'Passed':'Approved', 'M':'R', 'TV-G':'PG',
       'TV-PG':'PG'}
```

```python
data.content_rating = data.content_rating.map(conv)
```

```python
data.content_rating.unique()
```

### e. Actor and likes

```python
data.loc[data.actor_1_name.isna()]
```

```python
data.actor_1_name.fillna('None', inplace=True)
data.actor_2_name.fillna('None', inplace=True)
data.actor_3_name.fillna('None', inplace=True)
```

```python
likes = ['movie_facebook_likes', 'director_facebook_likes', 'actor_3_facebook_likes',
        'actor_1_facebook_likes', 'cast_total_facebook_likes', 'actor_2_facebook_likes']
review = ['num_critic_for_reviews', 'num_voted_users', 'num_user_for_reviews', 'imdb_score']
```

```python
data[likes].isna().sum()
```

```python
for column in likes:
    data.loc[:, column].fillna(0, inplace=True)
    
for column in review:
    data.loc[:, column].fillna(0, inplace=True)
```

```python
msno.bar(data)
```

### d. Gross

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10)) 
mcorr = data.corr()
sns.heatmap(mcorr, annot=True)
```

```python
No_null_gross = data[~data['gross' ].isnull()]
No_null_gross['gross'].isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

X = No_null_gross.loc[:, ['num_user_for_reviews', 'num_voted_users']]
Y = No_null_gross.gross
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4)
reg = LinearRegression()
reg.fit(X_train, Y_train)
Y_pred = reg.predict(X_test)
error = mean_absolute_error(Y_test, Y_pred)


print("Variance : " + str(round(r2_score(Y_test, Y_pred), 3)))
```

```python
data.loc[data.gross.isna(), 'gross'] = reg.predict(data.loc[data.gross.isna(),
                                                            ['num_user_for_reviews', 'num_voted_users']])
```

```python
msno.bar(data)
```

### 4. Saving CSV file

```python
data.to_csv('clean_datafile.csv', index=False)
```
