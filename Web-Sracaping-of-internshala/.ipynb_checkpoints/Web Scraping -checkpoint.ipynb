```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://internshala.com/internships/data%20science,machine%20learning-internship-in-delhi,ghaziabad,gurgaon,noida,faridabad,greater%20noida'
#conducting a request of the stated URL above:
page = requests.get(URL)
#specifying a desired format of “page” using the html parser - this allows python to read the various components of the page, rather than treating it as one long string.
soup = BeautifulSoup(page.text,"html.parser")
#printing soup in a more structured tree format that makes for easier reading
print(soup.prettify())
```

```python
lists = soup.find_all("div",{'class': "internship_meta"})
```

```python
list1=lists[1]
list1
```

```python
title=list1.find("h4").text
title[1:]
```

```python
loction=list1.find("a",{"class":"location_link"}).text
loction
```

```python
stipend=list1.find("td",{"class":"stipend_container_table_cell"}).text
stipend[:20]
```

```python
location = []
for div in lists:
    for a in div.find("a",{"class":"location_link"}):
        location.append(a)
location                   
```

```python
titles = []
for div in lists:
    for a in div.find("a"):
        titles.append(a)
titles   
```

```python
duration = []
for div in lists:
    for a in div.find_all("td")[1:2]:
        duration.append(a.text)
duration   
```

```python
stipend = []
for div in lists:
    for a in div.find_all("td")[2:3]:
        stipend.append(a.text[:20])
stipend[1:30]
```

```python
last_date = []
for div in lists:
    for a in div.find_all("td")[4:5]:
        last_date.append(a.text)
last_date   
```

```python
company_name = []
for div in lists:
    for a in div.find_all("a",{"class","link_display_like_text"}):
        company_name.append(a.text)
company_name
```

```python
link=[]
for links in soup.find_all("a",{'class': "view_detail_button"}):
    link.append("www.internshala.com"+links.get('href'))
    
```

```python
columns=[titles,location,duration,stipend,last_date]
```

```python
df = pd.DataFrame({"job Titles":titles,"Company Name":company_name,"Location":location,"Duration":duration,"Stipend":stipend,"Last date":last_date,"URL":link})
```
